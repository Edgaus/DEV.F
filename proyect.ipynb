```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df15=pd.read_csv("Violencia en Mexico/Datos completos/2015/2015_datos_violencia_delictiva.csv" )
df16=pd.read_csv("Violencia en Mexico/Datos completos/2016/2016_datos_violencia_delictiva.csv" )
df17=pd.read_csv("Violencia en Mexico/Datos completos/2017/2017_datos_violencia_delictiva.csv" )
df18=pd.read_csv("Violencia en Mexico/Datos completos/2018/2018_datos_violencia_delictiva.csv" )
df19=pd.read_csv("Violencia en Mexico/Datos completos/2019/2019_datos_violencia_delictiva.csv" )


df15.head()
```

```python
#Clasificacion de region de estados en Mexico
Noroeste=['Baja California','Baja California Sur','Chihuahua','Sinaloa','Sonora']
Noreste=['Coahuila de Zaragoza','Durango','Nuevo León','San Luis Potosí','Tamaulipas']
Centro=['Ciudad de México','México','Guerrero','Hidalgo','Morelos','Puebla','Tlaxcala']
Occidente=['Aguascalientes','Guanajuato','Colima','Jalisco','Nayarit','Michoacán de Ocampo','Querétaro','Zacatecas' ]
Sureste=['Campeche','Chiapas','Campeche','Veracruz de Ignacio de la Llave', 'Yucatán','Tabasco','Quintana Roo' ]

meses=list(df15.columns[9:21])
años=['2015','2016','2017','2018','2019']

#Definiendo columa total

df15['suma'],df16['suma'],df17['suma'],df18['suma'],df19['suma']=(0,0,0,0,0)   

#Sumando las columnas de los meses para obtener el total de delitos por año

for mes in meses :
    df15['suma']+=df15[mes]
    df16['suma']+=df15[mes]
    df17['suma']+=df15[mes]
    df18['suma']+=df15[mes]
    df19['suma']+=df15[mes]


#Vector con informacion para la grafica de información

condicion=['Entidad'=='Sinaloa']

#Violencia en las zonas Noroeste
noro_delitos=pd.DataFrame( {'2015': df15[df15.Entidad.isin(Noroeste)].loc[df15['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2016': df16[df16.Entidad.isin(Noroeste)].loc[df16['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2017': df17[df17.Entidad.isin(Noroeste)].loc[df17['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2018': df18[df18.Entidad.isin(Noroeste)].loc[df18['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2019': df19[df19.Entidad.isin(Noroeste)].loc[df19['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
}
)


#Violencia en las zonas Noreste

nore_delitos=pd.DataFrame( {'2015': df15[df15.Entidad.isin(Noreste)].loc[df15['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2016': df16[df16.Entidad.isin(Noreste)].loc[df16['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2017': df17[df17.Entidad.isin(Noreste)].loc[df17['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2018': df18[df18.Entidad.isin(Noreste)].loc[df18['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2019': df19[df19.Entidad.isin(Noreste)].loc[df19['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
}
)


#Violencia en las zonas Centro

cen_delitos=pd.DataFrame( {'2015': df15[df15.Entidad.isin(Centro)].loc[df15['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2016': df16[df16.Entidad.isin(Centro)].loc[df16['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2017': df17[df17.Entidad.isin(Centro)].loc[df17['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2018': df18[df18.Entidad.isin(Centro)].loc[df18['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2019': df19[df19.Entidad.isin(Centro)].loc[df19['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
}
)

#Violencia en las zonas Occidente


occ_delitos=pd.DataFrame( {'2015': df15[df15.Entidad.isin(Occidente)].loc[df15['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2016': df16[df16.Entidad.isin(Occidente)].loc[df16['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2017': df17[df17.Entidad.isin(Occidente)].loc[df17['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2018': df18[df18.Entidad.isin(Occidente)].loc[df18['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2019': df19[df19.Entidad.isin(Occidente)].loc[df19['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
}
)

#Violencia en las zonas Sur

sur_delitos=pd.DataFrame( {'2015': df15[df15.Entidad.isin(Sureste)].loc[df15['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2016': df16[df16.Entidad.isin(Sureste)].loc[df16['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2017': df17[df17.Entidad.isin(Sureste)].loc[df17['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2018': df18[df18.Entidad.isin(Sureste)].loc[df18['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
                           '2019': df19[df19.Entidad.isin(Sureste)].loc[df19['Tipo de delito']=='Feminicidio'].groupby('Entidad').agg(sum)['suma'],
}
)

```

```python
fig1, axes = plt.subplots(  nrows=1, ncols=3, figsize=(25, 5) )  


#Plot de los feminicidios en la region Noroeste

noro_delitos.T.plot(
    ax=axes[0], style=['o-b', '--r', ':k', '.-g','o-y'],
    
)

axes[0].legend(loc='lower center',bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True)

#Plot de los feminicidios en la region Noreste

nore_delitos.T.plot(
    ax=axes[1], style=['o-b', '--r', ':k', '.-g']
)

axes[1].legend(loc='lower center',bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True)

#Plot de los feminicidios en la region Occidente

occ_delitos.T.plot(
    ax=axes[2], style=['o-b', '--r', ':k', '.-g']
)

axes[2].legend(loc='lower center',bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True)



fig2, axes = plt.subplots(  nrows=1, ncols=2, figsize=(25, 5) )  


#Plot de los feminicidios en la region Centro

cen_delitos.T.plot(
    ax=axes[0], style=['o-b', '--r', ':k', '.-g']
)
axes[0].legend(loc='lower center',bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True)

#Plot de los feminicidios en la region Sureste

sur_delitos.T.plot(
    ax=axes[1], style=['o-b', '--r', ':k', '.-g']
)
axes[1].legend(loc='lower center',bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True)
```
